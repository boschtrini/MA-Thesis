```python
# Specify the path to your JSON file
json_file_path = '/Users/trinidadbosch/Desktop/SEDS/Tesis/Data/MA-Thesis/Google Patents/Data/all_words_freq.json'

# Read the JSON file and load its contents into a dictionary
with open(json_file_path, 'r') as json_file:
    patents = json.load(json_file)
```

```python
## Coger top 40k palabras que aparecen 50 veces en ambos corpus. Logaritmo de la frecuencia en corpus A - log B.
#FRecuencia reddit/patentes
## Top 40k palabras mas frequentes y calculo la interseccion, por cada palabra tengo 2 frecuencias R y Patentes
## log frecuancia A - log freq B. 
## centrarme en las barras moradas
```

```python
# Get an entropy shift
entropy_shift = rs.EntropyShift(type2freq_1=patents,
                                type2freq_2=reddit,
                                alpha = 1,
                                base=2)

entropy_shift.get_shift_graph(top_n = 200) 
```

```python
jsd_shift = rs.JSDivergenceShift(type2freq_1=reddit,
                                type2freq_2=patents,
                                base=2)

jsd_shift.get_shift_graph() 
```
